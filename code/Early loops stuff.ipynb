```python
rhos = pd.read_csv('autoc/total_s10by10.dat', header=None, skiprows=1).to_numpy()
# rhos = [rhos[5]]
print(str(np.shape(rhos)[0]) + " runs, each of " + str(np.shape(rhos)[1]-1) + " steps")

interval = np.shape(rhos)[1] - 1
cutoff=150

result = np.zeros((interval-cutoff)+1)
for rho in rhos:
    data = rho[cutoff:]
    x = data - np.mean(data)
    result += np.correlate(x, x, mode='full')[interval-cutoff:]/np.arange(1+interval-cutoff,0,-1)
result = result/len(rhos)

plt.plot(result)
# plt.ylim(-100,100)
# plt.savefig("img")
```

```python
cutoff=250
data = rhos[:,cutoff:]
interval = np.shape(data)[1]

result = np.zeros(interval)
for rho in data[50:]:
    partial_result = []
    x = rho - np.mean(rho)
    for t in range(interval):
        val = sum(x[t:] * x[:interval-t])/(interval-t)
        partial_result.append(val)
    result += partial_result
result /= np.shape(data)[0]
plt.plot(result)

result = np.zeros(interval)
for rho in data[:50]:
    partial_result = []
    x = rho - np.mean(rho)
    for t in range(interval):
        val = sum(x[t:] * x[:interval-t])/(interval-t)
        partial_result.append(val)
    result += partial_result
result /= np.shape(data)[0]
plt.plot(result)
```
